```python
# Import necessary libraries
import sys
sys.path.insert(0, '..')

import numpy as np
import librosa
from pathlib import Path

from src.audio_analyzer import AudioAnalyzer
from src.genre_classifier import GenreClassifier
from src.mood_detector import MoodDetector
from src.tag_generator import TagGenerator
```

## 1. Audio Loading and Feature Extraction

```python
# Initialize the analyzer
analyzer = AudioAnalyzer()

# Analyze a sample file
# Replace with your audio file path
sample_file = "sample.mp3"
```

## 2. Genre Classification

```python
# Initialize genre classifier
classifier = GenreClassifier()

# Test with simulated features
test_features = {
    'bpm': 128,
    'energy': 0.7,
    'spectral_centroid': 5000,
    'spectral_bandwidth': 3000,
    'spectral_contrast': 0.5,
    'valence': 0.6
}

genres = classifier.classify(test_features)
print("Genre classification results:")
for genre, confidence in genres:
    print(f"  {genre}: {confidence:.2%}")
```

## 3. Mood Detection

```python
# Initialize mood detector
mood_detector = MoodDetector()

moods = mood_detector.detect(test_features)
print(f"Detected moods: {moods}")

# Get mood vector
mood_vector = mood_detector.get_mood_vector(test_features)
print("\nMood intensity vector:")
for mood, intensity in mood_vector.items():
    if intensity > 0:
        print(f"  {mood}: {intensity:.2f}")
```

## 4. Tag Generation

```python
# Initialize tag generator
tag_gen = TagGenerator()

# Generate tags
tags = tag_gen.generate_tags(test_features)
print(f"Generated tags: {tags}")

# Get tag summary by category
summary = tag_gen.get_tag_summary(test_features)
print("\nTag summary:")
for category, category_tags in summary.items():
    print(f"  {category}: {category_tags}")
```

## 5. Feature Analysis Visualization

```python
import matplotlib.pyplot as plt

# Visualize genre profile ranges
genres_to_plot = ['electronic', 'house', 'hip_hop', 'ambient', 'rock']

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# BPM ranges
for genre in genres_to_plot:
    if genre in classifier.profiles:
        profile = classifier.profiles[genre]
        bpm_range = profile['bpm_range']
        axes[0].barh(genre, bpm_range[1] - bpm_range[0], left=bpm_range[0])

axes[0].set_xlabel('BPM')
axes[0].set_title('BPM Ranges by Genre')

# Energy ranges
for genre in genres_to_plot:
    if genre in classifier.profiles:
        profile = classifier.profiles[genre]
        energy_range = profile['energy_range']
        axes[1].barh(genre, energy_range[1] - energy_range[0], left=energy_range[0])

axes[1].set_xlabel('Energy')
axes[1].set_title('Energy Ranges by Genre')

plt.tight_layout()
plt.show()
```

## 6. Batch Processing Test

```python
# Test with multiple simulated audio features
test_tracks = [
    {'name': 'Track 1', 'features': {'bpm': 128, 'energy': 0.8, 'spectral_centroid': 6000, 'valence': 0.7}},
    {'name': 'Track 2', 'features': {'bpm': 85, 'energy': 0.3, 'spectral_centroid': 2500, 'valence': 0.4}},
    {'name': 'Track 3', 'features': {'bpm': 140, 'energy': 0.9, 'spectral_centroid': 7000, 'valence': 0.3}},
    {'name': 'Track 4', 'features': {'bpm': 95, 'energy': 0.4, 'spectral_centroid': 3500, 'valence': 0.6}},
]

print("Batch Processing Results:\n")
for track in test_tracks:
    name = track['name']
    features = track['features']
    
    genres = classifier.get_primary_genre(features)
    moods = mood_detector.detect(features)
    tags = tag_gen.suggest_tags(features, limit=5)
    
    print(f"🎵 {name}")
    print(f"   Genre: {genres}")
    print(f"   Mood: {', '.join(moods)}")
    print(f"   Tags: {', '.join(tags)}")
    print()
```
